```python
import pandas as pd
path = "./vgsales.csv"
df = pd.read_csv(path)
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
df.dropna(inplace= True)
```

```python
df
```

```python
model = LinearRegression()

x = df.iloc[:,6].values.reshape(-1,1)
y = df.iloc[:,10].values.reshape(-1,1)
#print(df.iloc[:,6].values.reshape(-1,1))
model.fit(x,y)
```

**df.iloc[:,6] => pour indexer la colonne NA_Sales , .values pour obtenir les valeur contenues dans la colonne, .reshape(-1,1)=> pour transformer en vecteur**

```python
#Import matplotlib to plot
import matplotlib.pyplot as plt
Y_pred = model.predict(x)
plt.scatter(x,y)
plt.plot(x,Y_pred, color ='red')
plt.show()

r_squared = model.score(x,y)
print("R^2 value is : ",r_squared)

b = model.intercept_
a = model.coef_

print("y = {} * x + {}".format(a[0][0],b[0]))

#or we can use
y = model.predict(x)
```
